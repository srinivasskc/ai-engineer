Unpack the Tuple Items


```python
developer = ('Alice', 34, 'Rust Developer')
```

```python
name, age, job = developer
```

```python
print(name)
print(age)
print(job)
```

```python
developer = ('Alice', 34, 'Rust Developer')
name, *rest = developer
```

```python
print(name)
print(rest)
```

Count() - how many times an item appears in a tuple.

```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust')
```

```python
programming_languages.count('Rust')
```

```python
programming_languages.count('JavaScript')
```

For count() - argument needs to be passed

```python
programming_languages.count()
```

index() -  to find the index where a particular item is present in a tuple

```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust')
programming_languages.index('Rust')
```

When index() cannot find the value - ValueError

```python
programming_languages.index('JavaScript')
```

index() - with value, start index

```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust', 'Python')
programming_languages.index('Python',3)
```

When Index() could not find value - ValueError


```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust', 'Python')
programming_languages.index('Python',3,5)
```

```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust', 'Python')
programming_languages.index('Python',1,5)
```

Sorted()

```python
numbers = (13, 2, 78, 3, 45, 67, 18, 7)
sorted(numbers)
```

```python
sorted(numbers,reverse=True)
```

```python
sorted(numbers,reverse=False)
```

```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust', 'Python')
sorted(programming_languages, key=len)
```
