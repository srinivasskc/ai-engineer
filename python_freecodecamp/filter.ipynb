Filter()

create a list starting from an existing iterable 

```python
words = ['tree', 'sky', 'mountain', 'river', 'cloud', 'sun']
words

def is_long_word(word):
    return len(word) > 4

long_words = list(filter(is_long_word,words))
print(long_words)
```

```python
long_cwords = [word for word in words if len(word)>4]
long_cwords
```
