While Loops


```python
secret_number = 3
guess_number = 0

while guess_number != secret_number:
    guess_number = int(input('Guess the number: (1-5): '))
    if guess_number != secret_number:
        print("Wrong! Try again..")

print("You got it!")
```

Break and Continue


Break: Iterate through the values and break out once the value is reached. Matching value is not returned.

```python
developer_names = ['Fred','Koen','Denis', 'Tom']

for developer in developer_names:
    if developer=="Koen":
        break
    print(developer)
```

Continue: Iterate through the values and Skips the value when the matching value is reached. Matching value is not returned.

```python
developer_names = ['Fred','Koen','Denis', 'Tom']

for developer in developer_names:
    if developer=="Koen":
        continue
    print(developer)
```

For-If-Else

```python
words = ['sky', 'apple', 'rhythm', 'fly', 'orange']

for word in words:
    for letter in word:
        if letter.lower() in 'aeiou':
            print(f"'{word}' contains vowel '{letter}'")
            break
    else:
        print(f"'{word}' has no vowels")
```

Ranges() - start is optional (starts with zero), stop is not inclusive. (stops at -1)

```python
for num in range(3):
    print(num)
```

```python
for num in range(1,5):
    print(num)
```

Increments by 1 - by default, we can change by step parameter.

```python
for num in range(1,10,2):
    print(num)
```

1 argument is mandatory in range()

```python
for num in range():
    print(num)
```

range() accepts only ints, not float

```python
for num in range(2.0):
    print(num)
```

Decremental Order - Negative for reverse order

```python
for num in range(40,0,-10):
    print(num)
```

To keep track of index element

```python
languages = ['Spanish', 'English', 'Russian', 'Chinese']
index = 0

for language in languages:
    print(f"{index} - {language}")
    index += 1
```

Easier way to do this by using enumerate(), 
keeps track of the index for an iterable and returns an enumerate object.

```python
languages = ['Spanish', 'English', 'Russian', 'Chinese']
enum = list(enumerate(languages))
enum
```

```python
languages = ['Spanish', 'English', 'Russian', 'Chinese']

for index, language in enumerate(languages):
    print(f"Index: {index} - Language: {language}")
```

By default index value is 0, we can set default value with optional parameter in enumerate

```python
languages = ['Spanish', 'English', 'Russian', 'Chinese']

for index, language in enumerate(languages,1):
    print(f"Index: {index} - Language: {language}")
```

If we want to iterate, multiple lists in parallel. We can use zip()

```python
developers = ['Naomi', 'Dario', 'Jessica', 'Tom']
ids = [1, 2, 3, 4]

list(zip(developers, ids))
```

For Loop with zip()

```python
developers = ['Naomi', 'Dario', 'Jessica', 'Tom']
ids = [1, 2, 3, 4]

for dev,id in zip(developers,ids):
    print(f"Name: {dev} \t ID: {id}")
```

Sum()

```python
numbers = [5, 10, 15, 20]
total = sum(numbers)
print(total) # Result: 50
```

sum() with start argument - sets initial value as 10. 

By default, initial value is 0.

10+5+10+15+20 = 60

```python
numbers = [5, 10, 15, 20]
total = sum(numbers, 10) # positional argument
print(total) # 60
```

sum() with keyword argument

keyword "start="

```python
numbers = [5, 10, 15, 20]
total = sum(numbers,start=10)
print(total) # 60
```
