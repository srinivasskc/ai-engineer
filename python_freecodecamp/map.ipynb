map() -  takes an iterable and applies a function to each of its elements.

```python
celsius = [0, 10, 20, 30, 40]

def celsius_to_fahrenheit(temp):
    return (temp * 9/5) + 32

fahrenheit = list(map(celsius_to_fahrenheit,celsius))
print(fahrenheit)
```
