Lists - Append()

```python
numbers = [1,2,3,4,5]
```

```python
numbers.append(6)
numbers
```

```python
even_numbers = [2,4,6,8,10]
numbers.append(even_numbers)
numbers
```

Extend() - Extending the numbers

```python
odd_numbers = [1,3,5,7,9]
odd_numbers2 = [11,13,15,17,19]
odd_numbers.extend(odd_numbers2)
odd_numbers
```

insert() - insert new element at index position

```python
numbers = [1,2,3,4,5]
numbers.insert(2,2.5)
numbers
```

remove() - remove element from the list

```python
numbers
numbers.remove(2.5)
numbers
```

remove() - only first occurrence of element


```python
numbers = [1,2,3,4,5,5,5,5]
numbers.remove(5)
numbers
```

pop() - remove element at specific location

```python
numbers = [1,2,3,4,5,5,5,5]
numbers.pop(5)
numbers
```

clear() - Clear the list

```python
numbers.clear()
numbers
```
