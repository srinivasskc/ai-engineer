Lambda Functions - for higher order functions like map(), filter()

Lambda Functions are called as "anonymous functions"

For single line expressions, lambda is good choice

Normal Function:

```python
def square(num):
    return num ** 2

print(square(4)) # 16
```

Same Example with lambda

```python
lambda num: num ** 2
```

Filter() to lambda()

```python
numbers = [1, 2, 3, 4, 5]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # [2, 4]
```

Convert list comprehension to lambda function

```python
result = [(num,'even') if num % 2 ==0 else (num,'odd') for num in range(21)]
print(result)
```

```python
lambda_result = lambda: [(num,'even') if num % 2 ==0 else (num,'odd') for num in range(21)]
result = lambda_result()
print(result)
```
